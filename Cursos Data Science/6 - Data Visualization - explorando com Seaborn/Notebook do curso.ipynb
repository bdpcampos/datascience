#### Teste de Hipótese

**H0 - A distribuição da gorjeta é igual em ambos os grupos**

**H1 - A distribuição da gorjeta é diferente entre os grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa_porcentagem = gorjetas.query('sobremesa == "Sim"').porcentagem
sobremesa_porcentagem
```

```python
sem_sobremesa_porcentagem = gorjetas.query('sobremesa == "Não"').porcentagem
sem_sobremesa_porcentagem
```

```python
u, p_valor = ranksums(sobremesa_porcentagem, sem_sobremesa_porcentagem)
[u, p_valor]
```

```python
print(f'P_valor é {p_valor * 100:.2f}%')
```

```python
p_valor <= 0.05 # Significancia de 5% ou Confianca de 95%
```

**Assim aceitamos H0, então a distribuição da gorjeta é igual em ambos os casos, quem pede sobremesa e quem não pede**
