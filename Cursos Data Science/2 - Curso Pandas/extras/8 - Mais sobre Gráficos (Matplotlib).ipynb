# Mais sobre Gráficos

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1) # Dentro da area tem 2 linhas, 2 colunas e esse que criamos esta na posicao 1
g2 = area.add_subplot(2, 2, 2) # Dentro da area tem 2 linhas, 2 colunas e esse que criamos esta na posicao 2
g3 = area.add_subplot(2, 2, 3) # Dentro da area tem 2 linhas, 2 colunas e esse que criamos esta na posicao 3
g4 = area.add_subplot(2, 2, 4) # Dentro da area tem 2 linhas, 2 colunas e esse que criamos esta na posicao 4
```

```python
g1.scatter(dados['Valor'], dados['Area'])
g1.set_title('Área X Valor')

g2.hist(dados['Valor'])
g2.set_title('Histograma Valor')

dados_g3 = dados['Valor'].sample(100) #pega uma amostra de tamanho X (100)
dados_g3.index = range(len(dados_g3))
g3.plot(dados_g3)
g3.set_title('Amostra Valores de Aluguel')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```

```python
area.savefig('dados/grafico.png', dpi = 300, bbox_inches = 'tight')
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 7))

dados = pd.read_csv('dados/aluguel_amostra.csv', sep = ';')
```

```python
dados
```

```python
area = plt.figure()
g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)
```

```python
grupo1 = dados.groupby('Tipo Agregado')['Valor']
label = grupo1.count().index
valores = grupo1.count().values

g1.pie(valores, labels = label, autopct='%1.1f%%')
g1.set_title('Total de Imóveis por Tipo Agregado')

grupo2 = dados.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values

g2.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Imóveis por Tipo')
```
