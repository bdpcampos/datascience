```python
media_iptu_por_m2 = dados['IPTU'] / dados['Area']

media_iptu_por_m2
```

```python
df_media = pd.DataFrame(media_iptu_por_m2, columns = ['Media'])
df_media
```

```python
def remove_outliers(df,columns,n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()
        
        df = df[(df[col] <= mean+(n_std*sd))]
        
    return df
```

```python
df_media['Media']
```

```python
nova_media = remove_outliers(df_media,['Media'], 3)
nova_media = nova_media.mean()
nova_media
```
